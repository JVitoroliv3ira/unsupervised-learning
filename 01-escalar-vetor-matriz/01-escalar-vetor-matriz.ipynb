### O que é um vetor?

- Um vetor é uma sequência ordenada de números. Ele pode representar muitas coisas:
    - Uma posição no espaço (ex: [x, y])
    - características de um dado (ex: [altura, peso, idade])
    - Probabilidades (ex: [0.1, 0.4, 0.6])
- Em álgebra linear, vetores são elementos de uma dimensão.
- No NumPy, vetores são arrays de 1 dimensão.

`````python
vetor = np.array([1, 2, 3])

print(f"Vetor: {vetor}")
print(f"Tipo: {type(vetor)}")
print(f"Dimensão (ndim): {vetor.ndim}")
print(f"Tamanho (len): {len(vetor)}")
print(f"Shape: {vetor.shape}")
`````

### Produto escalar com np.dot

O produto escalar (ou *dot product*) entre dois vetores de mesma dimensão resulta em um **escalar**.

#### Definição matemática

Dado dois vetores:

- a = [a₁, a₂, ..., aₙ]  
- b = [b₁, b₂, ..., bₙ]

O produto escalar é calculado como:

a · b = a₁·b₁ + a₂·b₂ + ... + aₙ·bₙ

Ou seja, é a soma das multiplicações dos elementos correspondentes.

#### Exemplo

Sejam:

- a = [1, 2, 3]  
- b = [4, 5, 6]

Então:

a · b = (1×4) + (2×5) + (3×6)  
a · b = 4 + 10 + 18 = **32**

#### Como o NumPy calcula isso

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(np.dot(a, b))  # Saída: 32

`````python
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

soma = v1 + v2
multiplicacao = v1 * v2
escalar = np.dot(v1, v2)

print(f"Soma: {soma}")
print(f"Multiplicação: {multiplicacao}")
print(f"Escalar: {escalar}")
`````

### O que é uma matriz

Uma **matriz** é uma tabela de números organizada em **linhas** e **colunas**.

Ela pode ser usada para representar dados, sistemas de equações, transformações lineares, imagens, entre outros.

#### Exemplo de matriz 2x3:

[
  [1, 2, 3],  
  [4, 5, 6]
]

Essa matriz tem:

- 2 linhas  
- 3 colunas  
- Forma (shape) = (2, 3)  
- Dimensão (ndim) = 2

#### Representação no NumPy:

```python
import numpy as np

A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
````

Cada elemento pode ser acessado por sua posição `[linha][coluna]`:

```python
A[0][1]  # retorna 2 (primeira linha, segunda coluna)
```

`````python
matriz = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(f"Vetor: {matriz}")
print(f"Tipo: {type(matriz)}")
print(f"Dimensão (ndim): {matriz.ndim}")
print(f"Tamanho (len): {len(matriz)}")
print(f"Shape: {matriz.shape}")
`````

`````python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print(f"Soma:\n{A + B}\n")

print(f"Subtração:\n{A - B}\n")

print(f"Multiplicação elemento a elemento:\n{A * B}\n")

print(f"Multiplicação matricial (np.dot):\n{np.dot(A, B)}\n")

print(f"Multiplicação matricial (com @):\n{A @ B}")
`````
