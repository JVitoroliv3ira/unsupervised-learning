### O que é um vetor?

- Um vetor é uma sequência ordenada de números. Ele pode representar muitas coisas:
    - Uma posição no espaço (ex: [x, y])
    - características de um dado (ex: [altura, peso, idade])
    - Probabilidades (ex: [0.1, 0.4, 0.6])
- Em álgebra linear, vetores são elementos de uma dimensão.
- No NumPy, vetores são arrays de 1 dimensão.

````python
vetor = np.array([1, 2, 3])

print(f"Vetor: {vetor}")
print(f"Tipo: {type(vetor)}")
print(f"Dimensão (ndim): {vetor.ndim}")
print(f"Tamanho (len): {len(vetor)}")
print(f"Shape: {vetor.shape}")
````

### Produto escalar

O produto escalar (ou **dot product**) entre dois vetores de mesma dimensão resulta em um **escalar**.

#### Definição matemática

Dados dois vetores \( \vec{a}, \vec{b} \in \mathbb{R}^n \):

\[
\vec{a} \cdot \vec{b} = \sum_{i=1}^{n} a_i b_i = a_1b_1 + a_2b_2 + \dots + a_nb_n
\]

#### Exemplo:

Sejam:

\[
\vec{a} = [1, 2, 3], \quad \vec{b} = [4, 5, 6]
\]

Então:

\[
\vec{a} \cdot \vec{b} = 1\cdot4 + 2\cdot5 + 3\cdot6 = 4 + 10 + 18 = \boxed{32}
\]

#### Como o NumPy calcula isso:

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

resultado = np.dot(a, b)
print(resultado)  # Saída: 32

````python
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

soma = v1 + v2
multiplicacao = v1 * v2
escalar = np.dot(v1, v2)

print(f"Soma: {soma}")
print(f"Multiplicação: {multiplicacao}")
print(f"Escalar: {escalar}")
````
